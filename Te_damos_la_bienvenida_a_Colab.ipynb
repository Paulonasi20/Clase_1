4.1 Suponga que un investigador utiliza datos sobre el tamaño de las clases ($TC$) y de los promedios de las calificaciones en los exámenes para 100 clases de tercer curso, para estimar la regresión MCO:

$$
\widehat{CalificaciónExamen} = 520,4 - 5,82 \times TC, R^2 = 0,08, ESR = 11,5.
$$

a) Un aula tiene 22 estudiantes. ¿Cuál es la predicción de la regresión para la calificación media en el examen para esa clase?

$$
\widehat{Y} = 520.4 -5.82(22)=392.36
$$

b) El año pasado, un aula tenía 19 estudiantes, y este año cuenta con 23 alumnos. ¿Cuál es la predicción de la regresión para la variación en la media de las calificaciones en el examen para la clase?  

$$
ΔTC=4⇒ΔY=−5.82×4=−23.28
$$

c) La media muestral del tamaño de la clase para 100 aulas es de 21,4. ¿Cuál es la media muestral de las calificaciones en el examen entre las 100 aulas? (Pista: repasar las fórmulas de los estimadores MCO).

$$
Yˉ=520.4−5.82⋅21.4=395.85
$$

d) ¿Cuál es la desviación típica muestral de las calificaciones en los exámenes entre las 100 aulas? (Pista: repasar las fórmulas de $R^2$ y del ESR).

PASO 1:
$$
SSR=11.52×98=12960.5
$$
PASO 2:
$$
SST=0.9212960.5​=14087.5
$$
PASO 3:
$$
sY​=9914087.5​
​=11.93
$$

```
import math

# Datos
intercepto = 520.4
beta = -5.82
R2 = 0.08
ESR = 11.5
n = 100

# a) Predicción TC = 22
TC_a = 22
Y_a = intercepto + beta * TC_a

# b) Cambio de 19 a 23
delta_TC = 23 - 19
delta_Y = beta * delta_TC

# c) Media muestral
TC_media = 21.4
Y_media = intercepto + beta * TC_media

# d) Desviación típica
SSR = (ESR**2) * (n - 2)
SST = SSR / (1 - R2)
desv_tipica = math.sqrt(SST / (n - 1))

# Resultados
print("a) Predicción (TC=22):", round(Y_a, 2))
print("b) Cambio en calificación:", round(delta_Y, 2))
print("c) Media de calificaciones:", round(Y_media, 2))
print("d) Desviación típica:", round(desv_tipica, 2))
```

4.2 Supóngase que se selecciona una muestra aleatoria de 200 varones de veinte años de edad de una población y que se registra la altura y el peso de estos hombres. Una regresión del peso sobre la altura da:

$$
\widehat{Peso} = -99,41 + 3,94 \times Altura, R^2 = 0,81, ESR = 10,2,
$$

donde $Peso$ se mide en libras y $Altura$ se mide en pulgadas.

a) ¿Cuál es la predicción que proporciona la regresión para el peso de alguien que mide 70 pulgadas? ¿y para alguien de 74 pulgadas?

70PULGADAS
$$
Peso^=−99.41+3.94(70)=176.39
$$

74PULGADAS
$$
Peso^=−99.41+3.94(74)=192.15
$$


b) Un hombre da un estirón tardío y crece 1,5 pulgadas a lo largo de un año. ¿Cuál es la predicción que proporciona la regresión para el aumento de peso de este hombre?

$$
ΔPeso=3.94×1.5=5.91
$$

c) Supóngase que en lugar de medir el peso y la altura en libras y pulgadas, esas variables se miden en centímetros y kilogramos. ¿Cuáles son las estimaciones de la regresión para esa nueva regresión centímetros-kilogramos? (Proporcione todos los resultados, los coeficientes estimados el $R^2$ y el $ESR$).

Cambio de cm a kg:

- 1 pulgada=2.54 cm
- 1 libra=0.453592 kg

Nuevo Coeficiente:
$$
β1∗​=2.543.94×0.453592​=0.703
$$


Nuevo intercepto:
$$
β0​=−99.41×0.453592=−45.09
$$

Nuevo ESR:

$$
ESR=10.2×0.453592=4.63
$$

Entonces:
$$
Peso^=−45.09+0.703⋅Altura
$$

- $R^2$=0.81 (no cambia)
- $ESR$=4.63



```
intercepto = -99.41
beta = 3.94
ESR = 10.2

# a) Predicciones
peso_70 = intercepto + beta * 70
peso_74 = intercepto + beta * 74

# b) Cambio
delta_peso = beta * 1.5

# c) Conversiones
lb_to_kg = 0.453592
in_to_cm = 2.54

beta_new = beta * lb_to_kg / in_to_cm
intercepto_new = intercepto * lb_to_kg
ESR_new = ESR * lb_to_kg

print("a) Peso 70 in:", round(peso_70, 2))
print("a) Peso 74 in:", round(peso_74, 2))
print("b) Cambio peso:", round(delta_peso, 2))
print("\nc) Nueva regresión (kg y cm):")
print("Intercepto:", round(intercepto_new, 2))
print("Pendiente:", round(beta_new, 3))
print("ESR:", round(ESR_new, 2))
print("R^2 = 0.81")
```
